# Positive, Negative or Zero

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

# Leap Year Checker

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

# Compare two numbers

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if num1 > num2:
    print("The first number is greater than the second number.")
elif num1 < num2:
    print("The first number is less than the second number.")
else:
    print("Both numbers are equal.")
```

# Range checker

```python
num = float(input("Enter a number: "))

if 10 < num < 100:  # Simpler way to write: num > 10 and num < 100
    print("Great!!")
else:
    print("Nah!!")
```

# Simple calculator

```python
num1 = float(input("Enter first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operator == '+':
    print(f"Result: {num1 + num2}")
elif operator == '-':
    print(f"Result: {num1 - num2}")
elif operator == '*':
    print(f"Result: {num1 * num2}")
elif operator == '/':
    if num2 != 0:
        print(f"Result: {num1 / num2}")
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operator.")
```

# Age Categorization

```python
age = int(input("Enter age: "))

if 0 <= age <= 12:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 64:
    print("Adult")
elif age >= 65:
    print("Senior")
else:
    print("Invalid age entered.")
```

# Classify a triangle

```python
side1 = float(input("Enter length of side 1: "))
side2 = float(input("Enter length of side 2: "))
side3 = float(input("Enter length of side 3: "))

if side1 == side2 == side3:
    print("Equilateral")
elif side1 == side2 or side2 == side3 or side1 == side3:
    print("Isosceles")
else:
    print("Scalene")
```

# String Comparism

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

if str1 == str2:
    print("Exactly equal")
else:
    print("Not exactly equal")
```

# Case Insensitive comparism

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

if str1.lower() == str2.lower():
    print("Same")
else:
    print("Different")
```

# Password strength criteria

```python
password = input("Enter your password: ")

if len(password) >= 8 and password != "password" and password != "123456":
    print("Password meets criteria.")
else:
    print("Password does not meet criteria.")
```

# Word Length classifier

```python
word = input("Enter a word: ")
length = len(word)

if 1 <= length <= 4:
    print("short")
elif 5 <= length <= 8:
    print("medium")
elif length >= 9:
    print("long")
else:
    print("Empty string")
```

# String content type

```python
text = input("Enter a string: ")

if text.isdigit():
    print("Contains only digits")
elif text.isalpha():
    print("Contains only letters")
elif text.isalnum():
    print("Contains a mix of both")
else:
    print("Contains spaces, symbols, or special characters")
```

# Single and multiple words

```python
text = input("Enter a string: ").strip()

if " " in text:
    print("Multiple words")
else:
    print("Single word")
```

# Vowel and Consonant checker

```python
name = input("Enter a name: ")

if name:  # Checks if string is not empty
    first_letter = name[0].lower()
    if first_letter in 'aeiou':
        print("Starts with a vowel")
    elif first_letter.isalpha():
        print("Starts with a consonant")
    else:
        print("Starts with a symbol or number")
else:
    print("Empty input")
```

# Sentence formatting checker

```python
sentence = input("Enter a sentence: ").strip()

if sentence:
    starts_capital = sentence[0].isupper()
    ends_punctuation = sentence[-1] in ['.', '!', '?']
    
    if starts_capital and ends_punctuation:
        print("Correct formatting")
    else:
        print("Incorrect formatting")
else:
    print("Empty sentence")
```

# Palindrome checker

```python
word = input("Enter a word: ").lower()

if word == word[::-1]:  # [::-1] reverses the string
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

# Phone number format checker

```python
phone = input("Enter phone number (XXX-XXX-XXXX): ")

# Check length and positions of dashes
if len(phone) == 12 and phone[3] == '-' and phone[7] == '-':
    # Remove dashes to check if everything else is a digit
    digits_only = phone.replace('-', '')
    if digits_only.isdigit():
        print("Valid format")
    else:
        print("Invalid format (must contain digits)")
else:
    print("Invalid format (check length and dashes)")
```
